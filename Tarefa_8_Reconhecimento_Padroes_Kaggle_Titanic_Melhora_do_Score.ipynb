```python
import pandas as pd
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, MaxAbsScaler
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, cross_validate, RepeatedKFold, KFold
import numpy as np
import datetime
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.metrics import fbeta_score, make_scorer, accuracy_score, precision_score
from sklearn.feature_selection import SelectKBest
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
X = train[list(test.columns)]
y = train[train.columns[~train.columns.isin(test.columns)]]
```

```python
def extraiPronome(nome):
    return nome.split(',')[1].split('.')[0].strip()

class AtributosDesejados(BaseEstimator, TransformerMixin):
    def __init__(self, excluirName=True):
        self.excluirName = excluirName
    def fit(self, X, y=None):
        self.colunasIndesejadas = ['PassengerId', 'Ticket', 'Cabin']
        if self.excluirName:
            self.colunasIndesejadas.append('Name')
        return self
    def transform(self, X, y=None):
        Xdrop = X.drop(self.colunasIndesejadas,axis=1)
        if 'Name' not in self.colunasIndesejadas:
            Xdrop['Name'] = Xdrop['Name'].apply(extraiPronome)
        return Xdrop
```

```python
class AtributosNumericos(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.colunasNumericas = X.select_dtypes(include='number').columns
        return self
    def transform(self, X, y=None):
        return X[self.colunasNumericas].to_numpy()
```

```python
class AtributosCategoricos(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.colunasCategoricas = X.select_dtypes(include='object').columns
        return self
    def transform(self, X, y=None):
        return X[self.colunasCategoricas].to_numpy()
```

```python
trataAtributos = Pipeline([
    ('unecaracteristicas', FeatureUnion([
        ('pipenum', Pipeline([
            ('atributos_numericos', AtributosNumericos()),
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ])),
        ('pipecat', Pipeline([
            ('atributos_categoricos', AtributosCategoricos()),
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('encoder', OneHotEncoder(handle_unknown='ignore'))
        ]))
    ])),
])
```

```python
# Modelo Professor Github
pipetotal = Pipeline([
    ('atributosDesejados', AtributosDesejados()),
    ('trataAtributos', trataAtributos),
    ('classificador', RandomForestClassifier())
])

parametros = {
    'atributosDesejados__excluirName': [True, False],
    'classificador__max_depth': [5]
}
modelo = GridSearchCV(pipetotal, param_grid=parametros)

```

```python
inicio = datetime.datetime.now()
scores = cross_validate(modelo, X, y, cv=RepeatedKFold())
score_ref = np.mean(scores['test_score'])
desvio_ref = np.std(scores['test_score'])
fim = datetime.datetime.now()
```

```python
# Referência
print(f'Start:{inicio}')
print(f'Start:{fim}')
print(f'Score Padrao:{score_ref}')
print(f'Desvio Ref:{desvio_ref}')
```

```python
# Tentativa de melhorar o score
trataAtributos_V = Pipeline([
    ('unecaracteristicas', FeatureUnion([
        ('pipenum', Pipeline([
            ('atributos_numericos', AtributosNumericos()),
            #('pca', PCA(2)),
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler_1', StandardScaler(copy=False, 
                                        with_mean=False,
                                       with_std=False)),
            ('scaler_2', MinMaxScaler(copy=False)),
            ('scaler_3', MaxAbsScaler())
        ])),
        ('pipecat', Pipeline([
            ('atributos_categoricos', AtributosCategoricos()),
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ('encoder', OneHotEncoder(handle_unknown='ignore'))
        ]))
    ])),
])
#ftwo_scorer = make_scorer(fbeta_score, beta=2)
#m_scorer = make_scorer(precision_score)
                       #greater_is_better=True)


```

```python

pipetotal_v = Pipeline([
    ('atributosDesejados', AtributosDesejados()),
    ('trataAtributos', trataAtributos_V),
    #('select', SelectKBest()),
    ('classificador', RandomForestClassifier())
])

parametros_v = {
    #'select__k': list(range(1,5,1)),
    'atributosDesejados__excluirName': [False],
    'classificador__n_estimators': [400],
    'classificador__max_depth': [10],
    'classificador__criterion': ['gini'],
    'classificador__min_samples_split': [10],
    'classificador__min_samples_leaf': [2],
    'classificador__max_features': ['auto'],
    'classificador__max_leaf_nodes': [20],
    'classificador__min_impurity_decrease': [0.0],
    'classificador__bootstrap': [False],
    'classificador__oob_score': [False],
    'classificador__n_jobs': [4],
    'classificador__random_state': [4],
    #'classificador__verbose': list(range(0,5,1))
    'classificador__warm_start': [True],
    #'classificador__class_weight': [None,'balanced', 'balanced_subsample'],
    #'classificador__ccp_alpha': [0.0,1.0,2.0],
    'classificador__max_samples': [1]
}
modelo_vinicius = GridSearchCV(pipetotal_v, 
                               param_grid=parametros_v, 
                               scoring='roc_auc_ovr_weighted',
                                  cv=3,
                               n_jobs=-1)
# scoring='roc_auc_ovr_weighted' - 0,871
# scoring='roc_auc' - 0,870
# scoring='roc_auc_ovr' - 0,869
# scoring='roc_auc_ovo' - 0,869
# scoring='roc_auc_ovo_weighted' - 0,868
# scoring='average_precision' - 0,852
# scoring='f1_micro'- 0,834
# scoring='accuracy' - 0,833
# scoring='f1_weighted'- 0,826
modelo_vinicius.fit(X, y)
```

```python
inicio_v = datetime.datetime.now()
scores_v = cross_validate(modelo_vinicius, X, y, cv=RepeatedKFold())
score_med_v = np.mean(scores_v['test_score'])
desvio_v = np.std(scores_v['test_score'])
fim_v = datetime.datetime.now()
```
